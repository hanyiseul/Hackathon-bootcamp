```python
%%writefile /content/drive/MyDrive/bootcamp/server.js

const express = require("express");
const path = require("path");
const app = express();

/* ============================================================
   1. 정적 파일(이미지, CSS, JS) 설정
   - 'static' 폴더 안에 있는 파일들을 "/static" 경로로 제공합니다.
   - 이제 HTML에서 이미지를 불러올 때 <img src="/static/logo.png"> 로 사용하세요.
   ============================================================ */
app.use("/static", express.static(path.join(__dirname, "static")));


/* ============================================================
   2. HTML 라우트 (templates 폴더 내 파일 연결)
   - 사용자의 요청에 따라 templates 폴더의 HTML을 전송합니다.
   ============================================================ */

const TEMPLATE_PATH = path.join(__dirname, "templates");

// [1] Home
app.get("/", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "/home.html"));
});

// [2] Dasboard
app.get("/dashboard", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "/dashboard.html"));
});

// [3] Map
app.get("/map", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "/map.html"));
});

/* ============================================================
   3. 서버 실행
   ============================================================ */
app.listen(4000, () => {
  console.log("Server is running on http://localhost:4000");
  console.log("HTML location: ./templates");
  console.log("Static location: ./static");
});
```
